# Analysis of Telco Churn Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/Churn_by_contract.csv')

sns.barplot(data=df, x='Contract', y='Churn_Rate')
plt.title('Churn Rate by Contract')
plt.xlabel('Contract')
plt.ylabel('Churn Rate')
plt.show()
```

This bar graph shows the churn rate by the type of contract the customer has. The Month-to month contract type has by far the largest Churn Rate. Two year contracts are very successful at retaining customers. Generally, the longer the contract, the more likely you are to retain the customer.

```python
df= pd.read_csv('/content/Churn_by_payment.csv')

sns.barplot(data=df, x='PaymentMethod', y='Churn_Rate')
plt.title('Churn Rate by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Here is the churn rate by the type of payment the customer used. Electronic check stands out as having a very high churn rate.

```python
df= pd.read_csv('/content/Churn_by_tenure.csv')

sns.barplot(data=df, x='tenure_bracket', y='Churn_Rate')
plt.title('Churn Rate by Tenure')
plt.xlabel('Tenure')
plt.ylabel('Churn Rate')
plt.show()
```

This bar graph shows the relationship between the tenure of the customer and their churn rate. Almost half of new customers are churned. The longer the tenure, the less likely the customers are to be churned.

```python
df= pd.read_csv('/content/Charge_by_if_Churned.csv')

sns.barplot(data=df, x='Churn', y='AverageMonthlyCharge')
plt.title('Average Monthly Charge for Churned vs Retained')
plt.ylabel('Average Monthly Charge')

plt.show()
```

Finally, we have a comparison between the monthly payments for the churned and retained customers. Churned customers payed \$70 on average while retained customers only payed \$60 per month.

# Analysis

The most important thing in retaining your customers is keeping them tenured for a long time. Try to sign new customers to two year contracts as those customers are much less likely to churn. Additionally, as they'll be tenured for two years, they'll be a long term customer, more likely to stay in the future.

Look into the electronic check payment method to see if there are any issues or difficulties for customers. Maybe ask customers who use that payment method for feedback to find out why it is disproportionately churned.

Finally, look into limiting the cost to a maximum so that cost doesn't result in losing customers. However, the difference between monthly payments for churned and retained customers may come from the higher average monthly price of the month-to-month contract type.
